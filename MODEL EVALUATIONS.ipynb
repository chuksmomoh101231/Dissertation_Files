**Comparing the base models with the stacked ensemble model**

```python
# Data
data = {
    "Metric": ["MAE", "MSE", "RMSE", "R2"],
    "LightGBM": [0.0083, 0.0097, 0.0987, 0.9993],
    "Random Forest": [0.0003, 0.0007, 0.0263, 0.9999],
    "Stacked Ensemble": [0.0003, 0.0004, 0.0204, 0.9999]
}

df = pd.DataFrame(data)

bar_width = 0.25

def plot_single_metric(metric):
    sub_df = df[df["Metric"] == metric]
    
    r1 = range(len(sub_df["LightGBM"]))
    r2 = [x + bar_width for x in r1]
    r3 = [x + bar_width for x in r2]

    plt.figure(figsize=(7, 5))

    plt.bar(r1, sub_df["LightGBM"], width=bar_width, label="LightGBM", color="b")
    plt.bar(r2, sub_df["Random Forest"], width=bar_width, label="Random Forest", color="r")
    plt.bar(r3, sub_df["Stacked Ensemble"], width=bar_width, label="Stacked Ensemble", color="g")

    # Add some text for labels, title, and custom x-axis tick labels, etc.
    plt.title(f"{metric} Comparison")
    plt.ylabel(metric, fontweight="bold")
    plt.xticks([r + bar_width for r in range(len(sub_df["LightGBM"]))], sub_df["Metric"])
    plt.legend()
    plt.tight_layout()

    plt.show()

# Plotting each metric separately
for metric in df["Metric"]:
    plot_single_metric(metric)

```

**Evaluation metrics for training and test dataset for Stacked Ensemble Model**

```python
# Data
data = {
    'Dataset': ['Hold-out', 'Training'],
    'MAE': [0.0003, 0.0002],
    'MSE': [0.0004, 0.0012],
    'RMSE': [0.0204, 0.0343],
    'R2': [0.9999, 0.9999]
}

df = pd.DataFrame(data)

# Plotting
fig, ax = plt.subplots(2, 2, figsize=(14, 10))

metrics = ['MAE', 'MSE', 'RMSE', 'R2']
positions = [(0, 0), (0, 1), (1, 0), (1, 1)]

for metric, pos in zip(metrics, positions):
    df.plot(x='Dataset', y=metric, kind='bar', ax=ax[pos[0], pos[1]], legend=False)
    ax[pos[0], pos[1]].set_title(metric)
    ax[pos[0], pos[1]].set_ylabel(metric)
    ax[pos[0], pos[1]].set_xticklabels(df['Dataset'], rotation=0)

plt.tight_layout()
plt.show()
```
